```python
def en_buyuk_sayi(sayilar):
  if not sayilar:
    return None
  en_buyuk = sayilar[0]
  for sayi in sayilar:
    if sayi > en_buyuk:
      en_buyuk = sayi
  return en_buyuk

sayilar=input("Sayıları virgülle ayırarak girin:").split(",")
sayilar=[int(sayi) for sayi in sayilar]
en_buyuk=en_buyuk_sayi(sayilar)
print(en_buyuk)
```
